# 读取excel、选择其中某列、条件筛选

```python
import pandas as pd
```

```python
data = pd.read_excel('./data/NHIS 2007 data.xlsx')
data
```

```python
new_bmi = data['weight'] * 0.45359237 / (data['height'] * 0.0254) ** 2
data['BMI'] - new_bmi
new_bmi
```

```python
weight = data[data['weight'] < 400]['weight'] * 0.45359237
height = data[data['height'] < 90]['height']  * 0.0254
bmi = weight / height ** 2
bmi
```

# 介绍如何用多个条件做数据筛选

```python
data[(data['weight'] < 400) & (data['height'] < 90)]
```

# 修改数据，保存到excel

```python
data['weight'] = data['weight'] * 0.45359237
data['height'] = data['height']  * 0.0254
data
```

```python
data.to_excel('./data/NHIS 2007 modified.xlsx')
```

```python
new_data = pd.read_excel('./data/NHIS 2007 modified.xlsx')
new_data
```

# 添加新列

```python
new_dmi = new_data['weight'] / new_data['height'] ** 2
new_data['new_dmi'] = new_dmi
new_data
```
